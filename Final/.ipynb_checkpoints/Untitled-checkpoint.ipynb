```python
import json
import re

# -----------------------------
# å·¥å…·å‡½æ•°ï¼šæ¸…ç†å­—ç¬¦ä¸²é‡Œçš„åˆ†æ•°ã€dash ç­‰
# -----------------------------

def normalize_text(s: str) -> str:
    if not isinstance(s, str):
        return s

    # Step 1: æŠŠ "2Â¾" è¿™ç§æ”¹æˆ "2 and Â¾"
    s = re.sub(r"(\d)(Â¼|Â½|Â¾|â…“|â…”)", lambda m: f"{m.group(1)} and {m.group(2)}", s)

    # Step 2: æŠŠåˆ†æ•°å­—ç¬¦æ›¿æ¢æˆæ ‡å‡†å†™æ³•
    s = (
        s.replace("Â¼", "1/4")
         .replace("Â½", "1/2")
         .replace("Â¾", "3/4")
         .replace("â…“", "1/3")
         .replace("â…”", "2/3")
         .replace("â€“", "-")
         .replace("â€”", "-")
         .strip()
    )

    return s

# -----------------------------
# å·¥å…·å‡½æ•°ï¼šæŠŠ "['a', 'b']" æˆ– "[\"a\"]" å˜æˆ Python list
# -----------------------------
def parse_array_string(value):
    if isinstance(value, list):
        return [normalize_text(v) for v in value]

    if isinstance(value, str):
        try:
            # æœ‰äº› JSON ä¼šç”¨å•å¼•å·
            json_str = value.replace("'", '"')
            data = json.loads(json_str)
            if isinstance(data, list):
                return [normalize_text(v) for v in data]
        except json.JSONDecodeError:
            # ä¸æ˜¯æ ‡å‡†çš„ JSON æ•°ç»„å­—ç¬¦ä¸²ï¼Œå°±æ‹†é€—å·
            return [normalize_text(v.strip()) for v in re.split(r",\s*", value) if v.strip()]
    return []

# -----------------------------
# ä¸»å‡½æ•°
# -----------------------------
def clean_json(input_file, output_file):
    with open(input_file, "r", encoding="utf-8") as f:
        data = json.load(f)

    cleaned = []
    for item in data:
        new_item = {}
        # ID â†’ id
        new_item["id"] = item.get("ID", item.get("id"))

        # Title â†’ title
        new_item["title"] = normalize_text(item.get("Title", item.get("title", "")))

        # Ingredients â†’ ingredients (è½¬æˆæ•°ç»„)
        new_item["ingredients"] = parse_array_string(item.get("Ingredients", []))

        # Instructions â†’ instructions
        new_item["instructions"] = normalize_text(item.get("Instructions", item.get("instructions", "")))

        # Image_Name â†’ imageName
        new_item["imageName"] = normalize_text(item.get("Image_Name", item.get("imageName", "")))

        # extracted_ingredients â†’ extractedIngredients
        new_item["extractedIngredients"] = parse_array_string(
            item.get("extracted_ingredients", item.get("extractedIngredients", []))
        )

        cleaned.append(new_item)

    # ä¿å­˜æˆæ–° JSON æ–‡ä»¶
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(cleaned, f, indent=2, ensure_ascii=False)

    print(f"âœ… æ¸…ç†å®Œæˆï¼å…±å¤„ç† {len(cleaned)} æ¡æ•°æ®ã€‚")
    print(f"ğŸ‘‰ è¾“å‡ºæ–‡ä»¶ï¼š{output_file}")


# -----------------------------
# è¿è¡Œ
# -----------------------------
if __name__ == "__main__":
    input_path = "Food_Ingredients.json"       # ä½ çš„åŸå§‹æ–‡ä»¶å
    output_path = "recipes_cleaned.json"  # è¾“å‡ºæ–‡ä»¶å
    clean_json(input_path, output_path)
```
